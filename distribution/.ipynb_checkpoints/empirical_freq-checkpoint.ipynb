```python
from sim_tools.distributions import (
    GroupedContinuousEmpirical,
)
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('bin.csv')
```

```python
group_dist =GroupedContinuousEmpirical(
    lower_bounds=data['lower_bound'],
    upper_bounds=data['upper_bound'],
    freq=data['y'],
    random_seed=None
)
```

```python
# Draw samples from the distribution
samples = group_dist.sample(size=10000)

plt.hist(samples, bins=50)
plt.xlabel("Inter-arrival time (hours)")
plt.ylabel("Frequency")
plt.title("Sampled Distribution")
plt.show()
```
